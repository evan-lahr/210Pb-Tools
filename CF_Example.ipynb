```python
from age_models import CF_age_model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#############################    INPUT CORE SPECIFIC CONSTANTS  ####################################
# values are transcribed from section 4.1, core information and constants.
Ψ          = 10              #cm         (internal diameter of core if circular)
u_Ψ        = 0.1             #cm         (uncertainty of internal diameter of core if circular)
S          = np.pi*(Ψ/2)**2  #cm^2       (cross sectional area of core if circular)
u_S        = np.pi*(Ψ/2)*u_Ψ #cm^2       (uncertainty of cross sectional area of core if circular)
C0         = 100.1           #Bq kg^-1   (concentration of 210Pb at the seabed Z=0)
u_C0       = 3               #Bq kg^-1   (uncertainty of concentration of 210Pb at the seabed Z=0)
T0_str     = '10/11/2004'    #MM/DD/YYYY (Sampling date AD)

###################################  MEASURED CORE DATA   ##########################################
core_data=pd.read_csv('cf_input.csv')

#run the age model
ages = CF_age_model(Ψ,   u_Ψ,    S,     u_S,     C0, u_C0,     T0_str,     core_data)
ages
```
